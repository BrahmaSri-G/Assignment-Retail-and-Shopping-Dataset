## Further Analysis & Data Visualization

### Product-Category wise Sales Volume

```python
Categories = df['Product Category'].unique()
print(Categories)
```

```python
Top_categories = df['Product Category'].value_counts()
print(Top_categories)
```

```python
Category_sales = df.groupby('Product Category')['Sales Volume'].sum().reset_index()
print(Category_sales.sort_values(by='Sales Volume', ascending=False))
```

```python
plt.figure(figsize=(6,4))
plt.bar(Category_sales['Product Category'], Category_sales['Sales Volume'], color='skyblue')
plt.title('Product category VS. Sales Volume')
plt.xlabel('Product Category')
plt.ylabel('Sales Volume')
plt.show()
```

### position wise Sales Volume

```python
position_sales = df.groupby('Product Position')['Sales Volume'].sum().reset_index()
print(position_sales.sort_values(by='Sales Volume', ascending=False))
```

```python
plt.figure(figsize = (6,4))
plt.bar(position_sales['Product Position'], position_sales['Sales Volume'], color=['lightgreen','lightblue','orange'])
plt.title('Product-Position Vs. Sales Volume')
plt.xlabel('Product Position')
plt.ylabel('Sales Volume')
plt.show()
```

### Product Promotion Vs. Sales Volume


```python
sns.barplot(x='Promotion', y='Sales Volume', data=df)
plt.title('Sales Volume by Promotion')
plt.show()
```

### Foot traffic Vs. Sales Volume

```python
sns.boxplot(x='Foot Traffic', y='Sales Volume', data=df, order=['Low', 'Medium', 'High'], hue='Foot Traffic',palette=['blue','orange','green'])
## plt.legend(title='Foot Traffic', loc='upper left')
plt.title('Sales Volume by Foot Traffic')
plt.show()
```

### Consumer Demographics Vs. Sales Volume

```python
df['Consumer Demographics'].unique()
```

```python
Consumer_demographics = df.groupby('Consumer Demographics')['Sales Volume'].sum()
print(Consumer_demographics)
```

```python
fig, ax=plt.subplots(figsize=(5,5))
ax.pie(Consumer_demographics.values, labels=Consumer_demographics.index, autopct='%1.1f%%', colors=['Skyblue','lightgreen', 'salmon','lightgrey'],startangle=90)
plt.title('Consumer Demographics')
plt.show()
```

## Seasonal effect on Sales Volume

```python
df['Seasonal'].unique()
```

```python
df['Seasonal'].value_counts()
```

```python
Seasonal_sales = df.groupby('Seasonal')['Sales Volume'].sum()
print(Seasonal_sales)
```

```python
Sales_year=df.groupby(['Seasonal','Product Category']).agg(Sales_year=('Price', 'sum')).reset_index()
print(Sales_year)
```

```python
Seasonal_sales.plot(kind='barh', title='Seasonal sales volume')
plt.xlabel('Sales Volume')
plt.xticks(rotation=45)
plt.ylabel('Seasonal')
plt.show()
```

## Price influence on Sales Volume

```python
sns.scatterplot(x='Price', y='Sales Volume', data=df, hue= 'Sales Volume')
plt.title('Price vs. Sales Volume')
plt.show()
sns.scatterplot(x='Competitor\'s Price', y='Sales Volume', data=df, hue= 'Sales Volume')
plt.title('Competitor\'s Price vs. Sales Volume')
plt.show()
```

```python
import plotly.express as px
# Aggregate sales volume by product category, product position, and promotion
agg_data = df.groupby(['Product Category', 'Product Position', 'Promotion']).sum().reset_index()

# Create sunburst chart
fig = px.sunburst(agg_data, path=['Product Category', 'Product Position', 'Promotion'], values='Sales Volume')
fig.update_layout(title_text='Sales Volume Distribution by Product Category, Position, and Promotion')
fig.show()
```
